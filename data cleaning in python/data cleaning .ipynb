```python
import pandas as pd  # Data manipulation and analysis library
import numpy as np  # Numerical computing library
```

```python
df = pd.read_csv(r"C:\Users\dices\Downloads\Nashville Housing Data for Data Cleaning (1).csv")
df.head()
```

```python
d_df = df.copy()
d_df.head()
```

```python
#standardize saledate
d_df['SaleDate'] = pd.to_datetime(d_df['SaleDate'], errors='coerce')
d_df['StandardizedSaleDate'] = d_df['SaleDate'].dt.strftime('%Y-%m-%d')
d_df.head()
```

```python
df.info()
```

```python
d_df['PropertyAddress'].isnull().sum()
```

```python
#populating property address 
d_df['FilledPropertyAddress'] = d_df.groupby('ParcelID')['PropertyAddress'].transform(lambda x: x.ffill().bfill())
d_df
```

```python
d_df['FilledPropertyAddress'].isnull().sum()
```

```python
#Breaking out Address into Individual Columns (Address, City, State)
d_df[['PropertyAddress_Street', 'PropertyAddress_City']] = d_df['PropertyAddress'].str.split(', ', expand=True)
d_df[['OwnerAddress_Street', 'OwnerAddress_City', 'OwnerAddress_State']] = d_df['OwnerAddress'].str.split(', ', expand=True)
d_df
```

```python
#Change Y and N to Yes and No in "Sold as Vacant" field

d_df['SoldAsVacant'].value_counts()
```

```python
d_df['SoldAsVacant'] = d_df['SoldAsVacant'].replace({'Y': 'Yes', 'N': 'No'})
d_df['SoldAsVacant'].value_counts()
```

```python
columns_to_drop = ['OwnerAddress', 'TaxDistrict', 'PropertyAddress', 'SaleDate']
d_df = d_df.drop(columns=columns_to_drop)
d_df
```

```python
#droping rows where 'OwnerName', 'Acreage', 'LandValue', 'BuildingValue', 'Bedrooms', 'FullBath', 'HalfBath', 'OwnerAddress_Street', 'OwnerAddress_City', 'OwnerAddress_State' are null

columns_to_check = ['OwnerName', 'Acreage', 'LandValue', 'BuildingValue', 'Bedrooms', 'FullBath', 'HalfBath', 'OwnerAddress_Street', 'OwnerAddress_City', 'OwnerAddress_State']
d_df = d_df.dropna(subset=columns_to_check, how='all')
d_df
```

```python
d_df.to_csv('your_updated_data.csv', index=False)
```
