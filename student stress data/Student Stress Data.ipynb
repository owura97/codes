```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\dices\\Downloads\\Compressed\\archive_3\\Student Stress Factors.csv")
df
```

```python
# Convert the 'Timestamp' column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True)
df
```

```python
df_r=df.rename(columns={"Kindly Rate your Sleep Quality üò¥":"Sleep Quality",
                   "How many times a week do you suffer headaches ü§ï?":"Headaches",
                  "How would you rate you academic performance üë©‚Äçüéì?":"Performance",
                  "how would you rate your study load?":"Study Load",
                  "How many times a week you practice extracurricular activities üéæ?":"Activities",
                  "How would you rate your stress levels?":"Stress Levels"})
df_r
```

```python
 df_r.describe()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))

for i, column in enumerate(df_r.columns[1:]):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df_r[column], kde=True)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df_r.corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
```

```python
variables_to_plot = df_r.columns[1:-1]
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))

for i, column in enumerate(variables_to_plot):
    plt.subplot(2, 3, i + 1)
    sns.barplot(x=column, y='Stress Levels', data=df_r)
    plt.title(f'{column} vs Stress Levels')

plt.tight_layout()
plt.show()
```

```python
# Crosstab analysis between Sleep Quality and Stress Level
crosstab_result = pd.crosstab(df_r['Sleep Quality'], df_r['Stress Levels'])

# Display the crosstab result
print("Crosstab Result:")
print(crosstab_result)
```

```python
sns.heatmap(crosstab_result, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.xlabel('Stress Levels')
plt.ylabel('Sleep Quality')
plt.title('Crosstab Analysis: Sleep Quality vs Stress Levels')
plt.show()
```
